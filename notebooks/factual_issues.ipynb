# Factual issues — random samples

This notebook uses the repository's shared setup and the notebook helpers to display:

- A random selection of factual issues from the entire corpus, and
- A random selection of up to 10 factual issues from each subject.

```python
# Load the standard dataset variables `issues` and `files`
%run _shared/setup_data_overview.py

# Load helper utilities
%run _shared/notebook_helpers.py
```

## Random factual issues from the whole corpus

```python
# Sample from the factual issues in the whole corpus
all_factual = issues[issues['Issue Category'].str.casefold() == 'factual'].copy()
all_factual = normalize_subjects(all_factual)

total_factual = len(all_factual)
sample_n = min(20, total_factual)

if total_factual == 0:
    display(Markdown("**No factual issues found in the corpus.**"))
else:
    sample_all = all_factual.sample(n=sample_n, random_state=42).reset_index(drop=True)
    display(Markdown(f"**Random sample of {sample_n} factual issues from the whole corpus (total={total_factual})**"))
    display_issues_table(sample_all)
```

## Random selection: up to 10 factual issues per subject

```python
# Sample up to 10 factual issues per subject
sample_per_subject = sample_factual_by_subject(issues, per_subject=10, random_state=42)

if sample_per_subject.empty:
    display(Markdown("**No factual issues by subject found.**"))
else:
    display(Markdown(f"**Displaying up to 10 factual issues from each subject — total displayed: {len(sample_per_subject)}**"))

    # Show counts per subject
    counts = sample_per_subject.groupby('Sampled Subject').size().reset_index(name='Count').sort_values('Count', ascending=False)
    display(Markdown('**Counts per subject:**'))
    display(counts)

    # Show sampled issues (subject, doc, page, context, issue, confidence, reasoning)
    display(Markdown('**Sampled issues:**'))
    display_issues_table(sample_per_subject, max_rows=500)
```

---

Tips:

- Change `sample_n` or the `per_subject` argument to adjust how many examples are shown.
- Use `display(sample_per_subject)` to get the raw DataFrame if you prefer that view.
